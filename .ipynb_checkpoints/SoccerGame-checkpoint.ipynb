```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
import matplotlib as mpl
mpl.rcParams['agg.path.chunksize'] = 10000
from datetime import datetime
```

```python
class SoccerGame:
    def __init__(self):
        # build a Soccer Env for 4 learners to run
        # Soccer Game
        # 0 1 2 3
        # 4 5 6 7
        self.A_pos = 2
        self.B_pos = 1
        #self.pos = np.random.choice([1,2,5,6],2,replace=False)
        #self.A_pos = self.pos[0]
        #self.B_pos = self.pos[1]
        #self.A_goal = [0, 4]
        #self.B_goal = [3, 7]
        # init score for Player A and B
        self.scores = [0, 0]
        # who has the ball(0:A, 1:B)
        #self.ball = np.random.randint(2)
        self.ball = 1
        # use a list to represent current state. [A_pos, B_pos, who_has_the_ball]
        self.state = [self.A_pos, self.B_pos, self.ball]
        # 0:North, 1:South, 2:East, 3:West, 4:Stick
        self.actions = [-4, 4, 1, -1, 0]
        self.done = False
        
    def check_action(self, player, action):
        # 0 for Player A, 1 for Player B
        if player == 0:
            temp_pos = self.A_pos
        else:
            temp_pos = self.B_pos
        # North
        if action == 0 and (temp_pos == 5 or temp_pos == 6):
            temp_pos += self.actions[action]
        elif action == 1 and (temp_pos == 1 or temp_pos == 2):
            temp_pos += self.actions[action]
        elif action == 2 and (temp_pos == 1 or temp_pos == 2 or temp_pos == 5 or temp_pos == 6):
            temp_pos += self.actions[action]
        elif action == 3 and (temp_pos == 1 or temp_pos == 2 or temp_pos == 5 or temp_pos == 6):
            temp_pos += self.actions[action]
        elif action == 4 and (temp_pos == 1 or temp_pos == 2 or temp_pos == 5 or temp_pos == 6):
            temp_pos =temp_pos
        return temp_pos
        
    def move(self, actions):
        if np.random.randint(2) == 0:
            # Player A move first
            temp_A_pos = self.check_action(0, actions[0])
            # collision condition
            if (temp_A_pos) == self.B_pos:
                # if A has the ball
                if self.ball == 0:
                    self.ball == 1
            # no collision
            else:
                self.A_pos = temp_A_pos
                # if A win
                if self.ball == 0:
                    if self.A_pos == 0 or self.A_pos == 4:
                        self.scores = [100, -100]
                        self.done = True
                    elif self.A_pos == 3 or self.A_pos == 7:
                        self.scores = [-100, 100]
                        self.done = True
                
            # Player B move secondly
            temp_B_pos = self.check_action(1, actions[1])
            # collision condition
            if (temp_B_pos) == self.A_pos:
                # if B has the ball
                if self.ball == 1:
                    self.ball == 0
            # no collision
            else:
                self.B_pos = temp_B_pos
                # if B win
                if self.ball == 1:
                    if self.B_pos == 3 or self.B_pos == 7:
                        self.scores = [-100, 100]
                        self.done = True
                    elif self.B_pos == 0 or self.B_pos == 4:
                        self.scores = [100, -100]
                        self.done = True
            self.state = [self.A_pos, self.B_pos, self.ball]
            return [self.state, self.scores, self.done]
        
        else:
            # Player B move first
            temp_B_pos = self.check_action(1, actions[1])
            # collision condition
            if (temp_B_pos) == self.A_pos:
                # if B has the ball
                if self.ball == 1:
                    self.ball == 0
            # no collision
            else:
                self.B_pos = temp_B_pos
                # if B win
                if self.B_pos == 3 or self.A_pos == 7:
                    self.scores = [-100, 100]
                    self.done = True
                elif self.B_pos == 0 or self.A_pos == 4:
                    self.scores = [100, -100]
                    self.done = True
            
            # Player A move secondly
            temp_A_pos = self.check_action(0, actions[0])
            # collision condition
            if (temp_A_pos) == self.B_pos:
                # if A has the ball
                if self.ball == 0:
                    self.ball == 1
            # no collision
            else:
                self.A_pos = temp_A_pos
                # if A win
                if self.A_pos == 0 or self.A_pos == 4:
                    self.scores = [100, -100]
                    self.done = True
                elif self.A_pos == 3 or self.A_pos == 7:
                    self.scores = [-100, 100]
                    self.done = True
            self.state = [self.A_pos, self.B_pos, self.ball]
            return [self.state, self.scores, self.done]
```

```python
# Q-Learning
print("Run Q-Learning")
```

```python
def Q_learning():
    gamma = 0.9             # discount factor
    epsilon = 1.0           # init epsilon
    epsilon_stop = 0.001   # final epsilon
    epsilon_decay = 0.99999 # decay rate
    alpha = 1.0             # learning rate
    alpha_stop = 0.001     # final lr
    alpha_decay = 0.99999
    max_iteration = 1000000      # terminate condition
    max_steps = 10000
    # Q table for Player A and B
    Q_A = np.zeros((8, 8, 2, 5))
    Q_B = np.zeros((8, 8, 2, 5))

    # epsilon greedy to select actions
    def epsilon_greedy(Q, state, epsilon):
        if np.random.random() >= epsilon:
            return np.argmax(Q[state[0]][state[1]][state[2]])
        else:
            return np.random.randint(0,5)
        
    # store the error term
    errors = []
    iteration = 0
    while iteration <= max_iteration:

        env = SoccerGame()
        done = False
        state_A = [env.state[0], env.state[1], env.state[2]]
        state_B = [env.state[1], env.state[0], env.state[2]]
        step = 0
        
        while step < max_steps:
            step += 1
            iteration += 1
            # generate action
            actions = [epsilon_greedy(Q_A, state_A, epsilon), epsilon_greedy(Q_B, state_B, epsilon)]
            
            # get next_state, rewards and done
            next_state, rewards, done = env.move(actions)
            next_state_A = [next_state[0], next_state[1], next_state[2]]
            next_state_B = [next_state[1], next_state[0], next_state[2]]
            
            # Q-learning for Player A and B
            old_Q_A = Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]]
            old_Q_B = Q_B[state_B[0]][state_B[1]][state_B[2]][actions[1]]
            # state as shown in fig.2, Player B has the ball, Player A move south
            old = Q_A[2][1][1][1]
            if not done:
                Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]] = (1-alpha)*Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]] + alpha *\
                (rewards[0] + gamma * np.max(Q_A[next_state_A[0]][next_state_A[1]][next_state_A[2]]))
                Q_B[state_B[0]][state_B[1]][state_B[2]][actions[1]] = (1-alpha)*Q_B[state_B[0]][state_B[1]][state_B[2]][actions[1]] + alpha *\
                (rewards[1] + gamma * np.max(Q_B[next_state_B[0]][next_state_B[1]][next_state_B[2]]))
                #errors.append(abs(Q_A[state_A[0]][state_A[1]][actions[0]] - old_Q_A))
                new = Q_A[2][1][1][1]
                errors.append(abs(new-old))
            if done:
                Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]] = (1-alpha)*Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]] + alpha * rewards[0]
                Q_B[state_B[0]][state_B[1]][state_B[2]][actions[1]] = (1-alpha)*Q_B[state_B[0]][state_B[1]][state_B[2]][actions[1]] + alpha * rewards[1]
                #errors.append(abs(Q_A[state_A[0]][state_A[1]][actions[0]] - old_Q_A))
                new = Q_A[2][1][1][1]
                errors.append(abs(new-old))
                break

            # update epsilon
            epsilon *= epsilon_decay
            epsilon = max(epsilon_stop, epsilon)
            # update alpha
            alpha *= alpha_decay
            alpha = max(alpha_stop, alpha)
            # update state
            state_A = next_state_A
            state_B = next_state_B
            #print("iteration:", iteration, epsilon, alpha, Q_A[1][0][3])
    return errors
```

```python
errors_QLearner = Q_learning()
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_axes([0.1,0.1,0.75,0.75])
ax.set_ylim(0,0.5)
ax.set_xlim(0,1e6)
ax.plot(errors_QLearner,ls='-',lw=1)
ax.set_title("Q-learner")
ax.set_xlabel("Simulation Iteration")
ax.set_ylabel("Q-value Difference")
fig.savefig("Q-Learner.png")
print("Save Q-Learner figure")
```

```python
def normalize(v):
    norm=np.linalg.norm(v, ord=1)
    if norm==0:
        norm=np.finfo(v.dtype).eps
    return v/norm
```

```python
def minimax(Q, state):
    a1, a2, a3, a4, a5, V = cp.Variable(),cp.Variable(),cp.Variable(),cp.Variable(),cp.Variable(),cp.Variable()
    constraints = [a1+a2+a3+a4+a5 == 1,
                  a1 >= 0,
                  a2 >= 0,
                  a3 >= 0,
                  a4 >= 0,
                  a5 >= 0,
                  sum(np.array(Q[state[0]][state[1]][state[2]].T[0]) * np.array([a1,a2,a3,a4,a5])) >= V,
                  sum(np.array(Q[state[0]][state[1]][state[2]].T[1]) * np.array([a1,a2,a3,a4,a5])) >= V,
                  sum(np.array(Q[state[0]][state[1]][state[2]].T[2]) * np.array([a1,a2,a3,a4,a5])) >= V,
                  sum(np.array(Q[state[0]][state[1]][state[2]].T[3]) * np.array([a1,a2,a3,a4,a5])) >= V,
                  sum(np.array(Q[state[0]][state[1]][state[2]].T[4]) * np.array([a1,a2,a3,a4,a5])) >= V]
    obj = cp.Maximize(V)
    prob = cp.Problem(obj, constraints)
    prob.solve()
    if a1.value<0 or a2.value<0 or a3.value<0 or a4.value<0 or a5.value<0:
        return normalize(np.array([abs(a1.value), abs(a2.value), abs(a3.value), abs(a4.value), abs(a5.value)])),V.value
    else:
        return [a1.value, a2.value, a3.value, a4.value, a5.value],V.value
```

```python
def Foe_Q():
    gamma = 0.9  # discount factor
    alpha = 1.0  # learning rate
    alpha_stop = 0.001  # final lr
    alpha_decay = 0.99999

    max_iteration = 1000000  # terminate condition
    max_steps = 10000
    # Q table for Player A and B, need to consider Player B's action and state
    Q_A = np.ones((8, 8, 2, 5, 5))  # 8 states for A, 8 states for B, who has the ball, 5 actions for A, 5 actions for B
    old = Q_A[2][1][1][1][4]
    prob = np.ones((8, 8, 2, 5)) * 0.2
    V = np.ones((8, 8, 2))


    # store the error term
    errors = []
    iteration = 0
    while iteration <= max_iteration:
        env = SoccerGame()
        done = False
        state_A = [env.state[0], env.state[1], env.state[2]]
        step = 0

        while step < max_steps:
            step += 1
            iteration += 1
            # generate action off policy
            actions = [np.random.randint(0, 5), np.random.randint(0, 5)]
            # get next_state, rewards and done
            next_state, rewards, done = env.move(actions)
            next_state_A = [next_state[0], next_state[1], next_state[2]]
            next_state_B = [next_state[1], next_state[0], next_state[2]]
            # Q-learning for Player A and B
            # state in fig 4 and Player A move South & Player B stick, B has the ball
            # do not keep updating old value. whenever in this state, update old value
            if state_A[0] == 2 and state_A[1] == 1 and state_A[2] == 1 and actions[0] == 1 and actions[1] == 4:
                old = Q_A[2][1][1][1][4]
                
            prob[state_A[0]][state_A[1]][state_A[2]],V[state_A[0]][state_A[1]][state_A[2]]= minimax(Q_A, state_A)
            if not done:
                Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]][actions[1]] = (1 - alpha) * \
                                                                                  Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]][actions[1]] + alpha * \
                                                                                  (rewards[0] + gamma * V[next_state_A[0]][next_state_A[1]][next_state_A[2]])
                new = Q_A[2][1][1][1][4]
                errors.append(abs(new - old))
            if done:
                Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]][actions[1]] = (1 - alpha) * \
                                                                                  Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]][actions[1]] + alpha * rewards[0]
                new = Q_A[2][1][1][1][4]
                errors.append(abs(new - old))
                break

            # update alpha
            alpha *= alpha_decay
            alpha = max(alpha_stop, alpha)

            # update state
            state_A = next_state_A
            state_B = next_state_B
            if iteration % 2000 == 0:
                print("iteration:", iteration,"error", errors[-1])
    return errors
```

```python
def CE(Q_A, Q_B):
    a = []
    for i in range(5):
        for j in range(5):
            if i != j:
                temp = np.zeros((5,5))
                temp[i,:] = Q_A[i] - Q_A[j]
                a.append(temp.reshape((25,)).tolist())
                
    for i in range(5):
        for j in range(5):
            if i != j:
                temp = np.zeros((5,5))
                temp[:,i] = Q_B[i] - Q_B[j]
                a.append(temp.reshape((25,)).tolist())
    prob = cp.Variable(25)
    a = np.array(a)
    u = (Q_A+Q_B.T).reshape(25)
    V = cp.Variable()

    constraints = [prob >= 0,
                  sum(prob) == 1,
                  a@prob >= 0,

                  V == u@prob 
       ]
    obj = cp.Maximize(V)
    problem = cp.Problem(obj, constraints)
    problem.solve()
    return prob.value.reshape((5,5))
```

```python
def CE_Q():
    gamma = 0.9  # discount factor
    alpha = 1.0  # learning rate
    alpha_stop = 0.001  # final lr
    alpha_decay = 0.99999
    max_iteration = 1000000  # terminate condition
    max_steps = 10000
    # Q table for Player A and B, need to consider Player B's action and state
    Q_A = np.ones((8, 8, 2, 5, 5))  # 8 states for A, 8 states for B, who has the ball, 5 actions for A, 5 actions for B
    Q_B = np.ones((8, 8, 2, 5, 5))  # 8 states for B, 8 states for A, who has the ball, 5 actions for B, 5 actions for A
    old = Q_A[2][1][1][1][4]
    V_A = np.ones((8, 8, 2))
    V_B = np.ones((8, 8, 2))

    # store the error term
    errors = []
    iteration = 0
    while iteration <= max_iteration:
        env = SoccerGame()
        done = False
        state_A = [env.state[0], env.state[1], env.state[2]]
        state_B = [env.state[1], env.state[0], env.state[2]]
        step = 0

        while step < max_steps:
            step += 1
            iteration += 1
            # generate action
            actions = [np.random.randint(0, 5), np.random.randint(0, 5)]
            # get next_state, rewards and done
            next_state, rewards, done = env.move(actions)
            next_state_A = [next_state[0], next_state[1], next_state[2]]
            next_state_B = [next_state[1], next_state[0], next_state[2]]
            # Q-learning for Player A and B
            # state in fig 4 and Player A move South & Player B stick, B has the ball
            # do not keep updating old value. whenever in this state, update old value
            if state_A[0] == 2 and state_A[1] == 1 and state_A[2] == 1 and actions[0] == 1 and actions[1] == 4:
                old = Q_A[2][1][1][1][4]
                
            # use CE function to get values 
            prob = CE(Q_A[state_A[0]][state_A[1]][state_A[2]], Q_B[state_B[0]][state_B[1]][state_B[2]])
            if prob is not None:
                V_A[state_A[0]][state_A[1]][state_A[2]] = np.sum(prob*Q_A[state_A[0]][state_A[1]][state_A[2]])
                V_B[state_B[0]][state_B[1]][state_B[2]] = np.sum(prob*Q_B[state_B[0]][state_B[1]][state_B[2]].T)
                #print(V_B[state_B[0]][state_B[1]][state_B[2]],'good')
            else:
                print("Solver fail at {}".format(iteration))
            if not done:
                Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]][actions[1]] = (1 - alpha) * \
            Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]][actions[1]] + alpha * \
            (rewards[0] + gamma * V_A[next_state_A[0]][next_state_A[1]][next_state_A[2]])
            
                Q_B[state_B[0]][state_B[1]][state_B[2]][actions[1]][actions[0]] = (1 - alpha) * \
            Q_B[state_B[0]][state_B[1]][state_B[2]][actions[1]][actions[0]] + alpha *\
            (rewards[1] + gamma * V_B[next_state_B[0]][next_state_B[1]][next_state_B[2]])
                new = Q_A[2][1][1][1][4]
                errors.append(abs(new - old))
            if done:
                Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]][actions[1]] = (1 - alpha) * \
            Q_A[state_A[0]][state_A[1]][state_A[2]][actions[0]][actions[1]] + alpha * rewards[0]
                Q_B[state_B[0]][state_B[1]][state_B[2]][actions[1]][actions[0]] = (1 - alpha) * \
            Q_B[state_B[0]][state_B[1]][state_B[2]][actions[1]][actions[0]] + alpha * rewards[1]
                new = Q_A[2][1][1][1][4]
                errors.append(abs(new - old))
                break

            # update alpha
            alpha *= alpha_decay
            alpha = max(alpha_stop, alpha)
            # update state
            state_A = next_state_A
            state_B = next_state_B
            if iteration % 2000 == 0:
                print("iteration:", iteration,"error", errors[-1],"time",datetime.now())
    return errors 
```
